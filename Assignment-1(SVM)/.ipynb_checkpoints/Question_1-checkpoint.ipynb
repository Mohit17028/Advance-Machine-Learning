```python
from sklearn.metrics import confusion_matrix
def confusion_mat_plot(Test_labels,acc_ovr):
    cm = confusion_matrix(Test_labels, acc_ovr)
    plt.matshow(cm)
    plt.title('Confusion matrix')
    plt.colorbar()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

def perf_measure(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
        if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
           FP += 1
        if y_actual[i]==y_hat[i]==0:
           TN += 1
        if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
           FN += 1
    return TP,FP,TN,FN
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
clf = LinearSVC(loss='hinge')
clf.fit(Train_data_A,Train_Label_A)
pre_A = clf.predict(Test_data_A)

acc= accuracy_score(pre_A, Test_Label_A)
confusion_mat_plot(Test_Label_A,pre_A)
TP,FP,TN,FN = perf_measure(Test_Label_A,pre_A)
print " TP,FP,TN,FN are",TP,FP,TN,FN
```

```python
#part D loss = hinged_square
import numpy as np
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
def Train_and_Fit(Data,Label,Test_data, Test_label):
    clf_t = LinearSVC(loss='squared_hinge')
    clf_t.fit(Data,Label)
    p =clf_t.predict(Test_data)
    plt.scatter(Test_data[:, 0], Test_data[:, 1], c=Test_label[:,0], s=30, cmap=plt.cm.Paired)

    # plot the decision function
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = clf_t.decision_function(xy).reshape(XX.shape)

    # plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    # plot support vectors
#     ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
#                linewidth=1, facecolors='none')
    plt.show()
    acc= accuracy_score(p,Test_label)
    return acc,p
acc_a, p_a=Train_and_Fit(Train_data_A,Train_Label_A,Test_data_A,Test_Label_A)
acc_b,p_b=Train_and_Fit(Train_data_B,Train_Label_B,Test_data_B,Test_Label_B)
acc_c,p_c=Train_and_Fit(Train_data_C,Train_Label_C,Test_data_C,Test_Label_C)
print "Accu_A  ",acc_a,"\n","Accu_ B  ",acc_b,"\n","Accu_C  ",acc_c,"\n"

```

```python
#Part E 
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
def Train_and_Fit(Data,Label,Test_data, Test_label):
    clf_t = SVC(kernel='rbf',probability=True)
    clf_t.fit(Data,Label)
    plt.scatter(Test_data[:, 0], Test_data[:, 1], c=Test_label[:,0], s=30, cmap=plt.cm.Paired)

    # plot the decision function
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = clf_t.decision_function(xy).reshape(XX.shape)

    # plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    # plot support vectors
#     ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
#                linewidth=1, facecolors='none')
    plt.show()
    p =clf_t.predict(Test_data)
    p_prob = clf_t.predict_proba(Test_data)
    acc= accuracy_score(p,Test_label)
    return acc, p,p_prob
acc_a_rbf, p_a_rbf,p_a_rbf_prob=Train_and_Fit(Train_data_A,Train_Label_A,Test_data_A,Test_Label_A)
acc_b_rbf,p_b_rbf,p_b_rbf_prob=Train_and_Fit(Train_data_B,Train_Label_B,Test_data_B,Test_Label_B)
acc_c_rbf,p_c_rbf,p_c_rbf_prob=Train_and_Fit(Train_data_C,Train_Label_C,Test_data_C,Test_Label_C)
```
